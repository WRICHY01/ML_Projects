```python
from typing import TypedDict, List, Annotated, Union
from langchain_core.messages import HumanMessage, AIMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.graph.message import add_messages
from langgraph.graph import StateGraph, START, END
from dotenv import load_dotenv, dotenv_values
```

```python
load_dotenv()
```

```python
class AgentState(TypedDict):
  message: Annotated[List[Union[HumanMessage]], add_messages]
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
def process(state: AgentState) -> AgentState:
  """
  A simple node that process and responds to human message
  """
  response = llm.invoke(state["message"])
  print(f"AI: {response.content}")

  return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("process_node", process)
graph.set_entry_point("process_node")
graph.set_finish_point("process_node")
agent = graph.compile()
```

```python
user_input = input("Enter Message: ")
while user_input != "quit":
  agent.invoke({"message": HumanMessage(content=user_input)})
  user_input = input("Enter Message")
```
