```python
from typing import List, TypedDict
import math
from langgraph.graph import StateGraph #Framework that aids in designing and managing the flow of task in the application using graph
```

```python
#AgentState creation: shared data structure that keeps track of information as the application runs

class AgentState(TypedDict):
  name: str
  values: List[int]
  operator: str
  result: str


def arithmetic_node(state: AgentState) -> AgentState:
  """
  Simple node that performs simple arithmetic
  """
  if state["operator"] == "*":
    # product = lambda x: x *= 
    state["result"] = f"Hi {state["name"]}, your answer is: {math.prod(state["values"])}"
  elif state["operator"] == "+":
    state["result"] = f"Hi {state["name"]}, your answer is: {sum(state["values"])}"
  else:
    print("wrong input from operator, use either '*' or '+' when inputting operator value")
    # return {"error_code":-1}
    pass

  return state
```

```python
#Creating a graph in langgraph
graph = StateGraph(AgentState)

graph.add_node("node_0", arithmetic_node)
# graph.add_node("node_0", greeting_node)

graph.set_entry_point("node_0")
graph.set_finish_point("node_0")

app = graph.compile()
# app?
```

```python
response = app.invoke({"name":"Jacqueline", "values": [3, 4, 6, 9], "operator": "-"})
response
```
